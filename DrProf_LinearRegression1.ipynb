# My first ML project

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

## * Data Preparation *

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis =1 )
x
```

### * Split Test Data *

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
x_test
```

## Model Testing

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### Applying the model to training set

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)

print(y_lr_train_pred, y_lr_test_pred)
```

#### Evaluate LR Model Performance

```python
y_train
```

```python
y_lr_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

```

```python
lr_train_mse, lr_test_mse
```

```python
lr_train_mse, lr_test_r2
```

```python
print('LR (Train) MSE', lr_train_mse)
print('LR (Train) R2', lr_train_r2)
print('LR (Test) MSE', lr_test_mse)
print('LR (Test) R2', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_test_mse, lr_train_r2, lr_test_r2]).transpose()

lr_results
```

```python
#lr_results.columns = ['Method', 'Training_MSE', 'Test_MSE', 'Training_R2', 'Test_r2']
```

```python
lr_results
```

### Random Forest

#### Trainings the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(x_train, y_train)

```

#### Evaluate the model

```python
from sklearn.metrics  import mean_squared_error, r2_score

y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF (Train) MSE', rf_train_mse)
print('RF (Train) R2', rf_train_r2)
print('RF (Test) MSE', rf_test_mse)
print('RF (Test) R2', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse,rf_test_mse,rf_train_r2,rf_test_r2]).transpose()
rf_results
```

```python
#rf_results.columns = ['Methods', 'Training_MSE', 'Test_MSE', 'Training_R2', 'Test_R2']
print(lr_results)
print(rf_results)
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results],axis = 0)
df_models
df_models.columns = ['Method', 'Training_MSE', 'Test_MSE', 'Training_R2', 'Test_R2']
df_models
```
